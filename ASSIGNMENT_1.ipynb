```
from sympy import *
b=symbols('b')
f=b*4+3*b*2+10
s=diff(f,b) 
print(s)
B=int(input("enter b value:"))
itr=1 
n=0.1 
epoch=2
while (itr <= epoch) :
   S=s.subs(b,B)
   print("s",S)
   deltax=-n*S
   print("delta(b) =",deltax)
   newx = B + deltax
   print("b =",newx)
   itr= itr + 1
   B=newx
print("final x =",B)
print("Minimum value is",f.subs(b,B))
```
